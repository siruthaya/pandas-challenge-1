## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
def sub_total(row):  
    return  row['unit_price'] * row['qty']
 
df['row_subtotal'] = df.apply(lambda row: sub_total(row), axis=1)

print(df[['unit_price', 'qty', 'row_subtotal']].head(10))
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def ship_price(row):  
    item_weight = row['unit_weight'] * row['qty']
    if item_weight > 50:
        shipping_price = item_weight * 7
    else:
        shipping_price = item_weight *10
    return shipping_price 
df['shipping_price'] = df.apply(lambda row: ship_price(row), axis=1)

print(df[['unit_price', 'unit_weight', 'qty', 'shipping_price']].head(10))
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
def tot_price(row):  
    item_price = row['row_subtotal'] + row['shipping_price'] + ((row['row_subtotal'] + row['shipping_price']) * .0925) 
    return round(item_price  ,2) 

df['total_price'] = df.apply(lambda row: tot_price(row), axis=1)

print(df[['row_subtotal', 'shipping_price', 'total_price']].head(10))
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

def cal_cost(row):  
    cost = (row['unit_cost'] * row['qty'])+ row['shipping_price'] 
    return round(cost,2) 
df['cost'] = df.apply(lambda row: cal_cost(row), axis=1)
df.head(10)
```

```python
# Create a column for the profit of each line using line cost and line price
def cal_profit(row):  
    profit = (row['total_price'] - row['cost'])
    return profit 

df['profit'] = df.apply(lambda row: cal_profit(row), axis=1)
df.head(10)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

for top_client in  df["client_id"].value_counts().head(5).index.tolist():
    print(top_client,':',f"{round(df.loc[df['client_id'] == top_client, 'total_price'].sum(),2):}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

summary_df = df.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('total_price', 'sum'),
    total_revenue=('cost', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()
top_clients_df = summary_df.sort_values(by='total_revenue', ascending=False).head(5)
top_clients_df
```

```python
# Format the data and rename the columns to names suitable for presentation.
top_clients_df = top_clients_df.rename(columns={'client_id':'Client ID', \
                                                'total_units_purchased':'Units',\
                                                'total_shipping_price':'Shipping (millions)',\
                                                    'total_revenue':'Total Revenue (millions)',\
                                                        'total_profit':'Total Profit (millions)'})
# Define the money columns. 

# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    return value / 1000000


# Apply the currency_format_millions function to only the money columns. 
top_clients_df['Shipping (millions)'] = top_clients_df['Shipping (millions)'].apply(currency_format_millions)
top_clients_df['Total Revenue (millions)'] = top_clients_df['Total Revenue (millions)'].apply(currency_format_millions)
top_clients_df['Total Profit (millions)'] = top_clients_df['Total Profit (millions)'].apply(currency_format_millions)


# Rename the columns to reflect the change in the money format. 
#top_clients_df = top_clients_df.rename(columns={
#    'Shipping (millions)': 'Shipping (in millions)',
 #   'Total Revenue (millions)': 'Total Revenue (in millions)',
 #   'Total Cost (millions)': 'Total Cost (in millions)'
#})

top_clients_df.head(10)

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
profit_sorted_df = top_clients_df.sort_values(by='Total Profit (millions)', ascending=False) 
profit_sorted_df.head(10)
```
